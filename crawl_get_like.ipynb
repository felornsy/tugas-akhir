```python
df = pd.read_csv('data/text-preprocessing.csv')
df = df.drop(columns=['Unnamed: 0'])
df.head(5)
```

```python
client = tweepy.Client(bearer_token)
client_paid = tweepy.Client(bearer_token_paid)
client_paid_1 = tweepy.Client(bearer_token_paid_1)
```

```python
loop = 1
tweetpy_client = client
for index, row in df.iterrows():
    
    if loop == 300:
        tweetpy_client = client_paid
    elif loop == 600:
        tweetpy_client = client_paid_1
    elif loop == 900:
        loop = 0
        time.sleep(60*15)
        tweetpy_client = client
    
    loop+=1
    practice_tweets = tweetpy_client.get_tweet(
        id = int(row['tweet_id']) ,
        tweet_fields = ["attachments","created_at", "entities", "geo", "id","public_metrics","possibly_sensitive", "source", "text", "author_id"],
    )
    if practice_tweets.data:
        like = practice_tweets.data.public_metrics["like_count"]
        df.loc[index, 'like_count'] = like
```
